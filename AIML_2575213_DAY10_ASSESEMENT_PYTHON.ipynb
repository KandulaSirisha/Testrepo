# 1.What are the ten most common movie names of all time?

```python
# Group the 'titles' DataFrame by movie title and count occurrences
common_names = titles['title'].value_counts().reset_index()
# Rename the columns for clarity
common_names.columns = ['Movie Title', 'Count']
# Sort the results in descending order by count
common_names = common_names.sort_values(by='Count', ascending=False)
top = common_names.head(10)
print(top)
```

# 2.Which three years of the 1930s saw the most films released?

```python
# Filter the 'titles' DataFrame to select only films from the 1930s
films = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]
# Group the filtered DataFrame by release year and count the number of films in each year
film_counts = films['year'].value_counts().reset_index()
# Rename the columns for clarity
film_counts.columns = ['Year', ' Film Count']
# Sort the results in descending order by film count
film_counts = film_counts.sort_values(by=' Film Count', ascending=False)
# Select the top three years with the most films released
top_three_years = film_counts.head(3)
print(top_three_years)
```

# 3.Plot the number of films that have been released each decade over the history of cinema.

```python
import matplotlib.pyplot as plt
# Filter the 'titles' DataFrame to select only films with valid years
valid_years = titles[titles['year'].notna()]

# Create a new column 'decade' to group films by decade
valid_years['decade'] = (valid_years['year'] // 10) * 10

# Group the DataFrame by decade and count the number of films in each decade
film_counts_by_decade = valid_years['decade'].value_counts().reset_index()

# Rename the columns for clarity
film_counts_by_decade.columns = ['Decade', 'Film Count']

# Sort the results by decade in ascending order
film_counts_by_decade = film_counts_by_decade.sort_values(by='Decade')

# Plot the number of films released each decade
plt.figure(figsize=(4, 2))
plt.bar(film_counts_by_decade['Decade'], film_counts_by_decade['Film Count'], width=5, align='edge')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(range(1890, 2030, 10), rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4.Plot the number of "Hamlet" films made each decade.

```python
import matplotlib.pyplot as plt
# Filter the DataFrame for movies titled "Hamlet"
hamlet_movies = titles[titles['title'] == 'Hamlet'].copy()  # Create a copy of the DataFrame

# Calculate the decade for each "Hamlet" film using .loc
hamlet_movies['decade'] = (hamlet_movies['year'] // 10) * 10

# Group the data by decade and count the number of "Hamlet" films in each decade
hamlet_decade_counts = hamlet_movies['decade'].value_counts().sort_index()

# Plot the number of "Hamlet" films made each decade
plt.figure(figsize=(3, 2))
plt.bar(hamlet_decade_counts.index, hamlet_decade_counts.values, width=5, align='center')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films')
plt.title('Number of "Hamlet" Films Made Each Decade')
plt.xticks(hamlet_decade_counts.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 8.Build a plot with a line for each rank n=1 through n=3, where the line shows what fraction of that rank's roles were 'actor' roles for each year in the history of film.

```python
# Define the rank values to include in the plot (n=1, n=2, n=3)
rank_values = [1, 2, 3]

# Initialize a list to store the line colors for each rank
line_colors = ['blue', 'green', 'red']

# Initialize a dictionary to store the DataFrames for each rank
rank_dataframes = {}

# Iterate over the rank values and create a DataFrame for each rank
for rank in rank_values:
    # Filter the 'cast' DataFrame to select only rows with the specific rank
    rank_df = cast[cast['n'] == rank]
    
    # Group the filtered DataFrame by 'year' and 'type' and count the number of roles for each year and type
    rank_by_year = rank_df.groupby(['year', 'type'])['character'].count().unstack()
    
    # Fill missing values (if any) with 0
    rank_by_year = rank_by_year.fillna(0)
    
    # Calculate the fraction of 'actor' roles each year
    total_roles_each_year = rank_by_year['actor'] + rank_by_year['actress']
    fraction_actor_roles = rank_by_year['actor'] / total_roles_each_year
    
    # Store the DataFrame in the dictionary
    rank_dataframes[rank] = fraction_actor_roles

# Plot the lines for each rank
plt.figure(figsize=(12, 6))
for i, rank in enumerate(rank_values):
    # Get the DataFrame for the current rank
    rank_df = rank_dataframes[rank]
    
    # Plot the line for the rank with a specific color
    plt.plot(rank_df.index, rank_df, label=f'n={rank}', color=line_colors[i], linestyle='-', marker='o')

plt.xlabel('Year')
plt.ylabel('Fraction of Actor Roles')
plt.title('Fraction of Actor Roles by Rank (n=1 to n=3) Each Year in Film History')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.ylim(0, 1)  # Set the y-axis limits to 0 to 1
plt.show()
```
